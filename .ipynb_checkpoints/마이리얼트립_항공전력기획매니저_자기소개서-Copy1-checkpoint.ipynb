## 자기소개 <br>


### 1. 마이리얼트립에 대해 알게 된 계기
<hr></hr>
 
처음 마이리얼트립을 접하게 된 계기는 SNS 광고와 실시간 검색어(항공 프로모션)를 통해서였습니다. 당시에는 세련된 여행업을 운영하는 스타트업이라고 생각하였고, 최근에는 채용사이트의 채용게시글을 통해 접하였습니다. 주요업무와 지원자격 등 Job Description을 보고 지원을 고민 하던 중, 평소 즐겨듣는 "데이터홀릭"이라는 팟캐스트 방송에 마이리얼트립 양승화 팀장님께서 출연하여 마이리얼트립은 어떤 회사이며, 어떤 사람을 원하는지에 대해 설명하는것을 들었습니다. 설명을 들은 후에는 "이 회사에 내가 필요할 이유"와 "내가 이 회사가 필요한 이유"에 대한 정리를 명확히 할 수 있었습니다. 

<br>

### 2. 마이리얼트립에 내가 필요할 이유
<hr></hr>

**<q> 능동적이며 성과에 기여할 수 있는 </q>**

현재의 급격한 트렌드 변화속에서 빠르게 적응할 수 있는 저만의 "핵심역량"을 찾고 갖추기 위하여 스스로 노력하여 왔습니다. 앞으로 기업이 매력적으로 느낄 인재는 무엇인가에 대한 고민을 많이 하였고, 저는 "데이터 기반 핵심역량"을 무기로 변화에 빠르게 적응할 수 있는 인재가 되고자 하였습니다. 대학 전공수업부터 외부교육,세미나,mooc,데이터 분석 프로젝트 등 다양한 방면으로 "데이터 skill" 향상을 위한 노력과  "데이터 인사이트"를 어떻게 비즈니스 이슈로 연결시킬것인가에 대한 고민 또한 해왔습니다. 

마이리얼트립과 같이 급격히 성장하는 기업의 경우, 데이터에 근거한 의사결정이 더욱 필요하다고 생각합니다. 저는 그 동안의 교육과 경험을 바탕으로 데이터 분석력을 키웠습니다. 데이터 분석과 이해를 통해 데이터에 근거한 의사결정에 기여하고 그로스팀,마케팅 부서와 효과적으로 커뮤니케이션을 통해 마이리얼트립의 성장에 기여하겠습니다.

<br>

### 3.  마이리얼트립이 나에게 필요한 이유
<hr></hr>

**<q> 같이, 또 빠르게 성장할 수 있는 </q>**

마이리얼트립 관련 뉴스와 인터뷰, 팟캐스트 방송을 들으며 느낀 마이리얼트립의 조직문화는 "같이, 또 빠르게 성장" 할 수 있는 기회와 환경을 제공해 준다는 것입니다. 사내 스터디에 구성원들이 열정적으로 참여한다는것과 데이터에 대한 자유로운 접근 또, 낮은 연차에도 주도적으로 프로젝트를 리드해 나갈 수 있다는 말은, 구성원 모두 개인과 회사의 성장을 바란다고 생각하였습니다.

마이리얼트립과 함께 제 개인의 성장을 이룸과 동시에 마이리얼트립의 발전에 기여하고 싶습니다.
