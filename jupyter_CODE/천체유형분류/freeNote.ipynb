```python
import pandas as pd
```

```python
df = pd.read_csv('submission_simpleLGBM.csv')
guide = pd.read_csv('guide_submit.csv')
```

```python
copy = df.copy()
```

```python
each_row_maxcol = df.iloc[:,1:].idxmax(axis=1)

for i,x in enumerate(each_row_maxcol):
    max_val = df.iloc[i,1:].max()
    df.iloc[i] = 0
    df.loc[i,x] = max_val
```

```python
df['id'] = guide['id']
```

```python
df.to_csv('submission_simpleLGBM_2.csv',index=False)
```
