```python
import pandas as pd
import numpy as np
df_t = pd.read_csv('teamplay.csv')
```

```python
df_final = pd.read_csv('final.csv')
df_final['ZIP_CODE'] =df_t['ZIP']
```

```python
df_a = df_final[['MAJOR','TOTAL','INCOME','NUMCHLD','AGE','WEALTH1','GENDER','ZIP_CODE']]
df_a[['CARDPROM','NUMPROM','STATE','ZIP_CODE']]=df_t[['CARDPROM','NUMPROM','STATE','ZIP']]

df_a[['INCOME','NUMCHLD','WEALTH1','ZIP_CODE']]=df_a[['INCOME','NUMCHLD','WEALTH1','ZIP_CODE']].astype('category')
df_a.to_csv('ddddddddd.csv',index=0)
```

```python
## STATE 별 MAJOR 분포 코드
df_filterd_state_top_CA=df_filterd[df_filterd['STATE']=='CA']
df_filterd_state_top_FL=df_filterd[df_filterd['STATE']=='FL']
df_filterd_state_top_TX=df_filterd[df_filterd['STATE']=='TX']

df_filterd_state_top_NV=df_filterd[df_filterd['STATE']=='NV']
df_filterd_state_top_NM=df_filterd[df_filterd['STATE']=='NM']
df_filterd_state_top_AR=df_filterd[df_filterd['STATE']=='AR']
```

```python
## 기부액 상위 3개 주 추출 ,, 한글패치
df_state_top = pd.concat([df_filterd_state_top_CA,df_filterd_state_top_FL,df_filterd_state_top_TX])


df_send=df_state_top.rename(columns ={'MAJOR' : '고액기부자','STATE' : '주' ,'AGE' : '나이',
                                      'GENDER' : '성별' ,'NUMCHLD' : '자녀수','WEALTH1' :'부의등급','INCOME' :'소득',
                                      'CARDPROM' :'카드프로모션','NUMPROM':'프로모션총횟수'})
```

```python
## 기부액 중간 3개 주 추출
df_state_middle = pd.concat([df_filterd_state_top_NV,df_filterd_state_top_NM,df_filterd_state_top_AR])

df_send2=df_state_middle.rename(columns ={'MAJOR' : '고액기부자','STATE' : '주' ,'AGE' : '나이',
                                      'GENDER' : '성별' ,'NUMCHLD' : '자녀수','WEALTH1' :'부의등급','INCOME' :'소득',
                                      'CARDPROM' :'카드프로모션','NUMPROM':'프로모션총횟수'})

df_send2.to_csv('abcdefg2.csv',index=0)

```
