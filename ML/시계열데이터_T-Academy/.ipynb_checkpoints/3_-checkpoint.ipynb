## SMA, WMA, SES 모델

#### Simple Moving Average

* 이동평균
* window size만큼의 평균을 구하여 예측

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
airline =  pd.read_csv('./data/airline_passengers.csv', index_col='Month',parse_dates=True)
```

```python
airline.head()
```

```python
airline['6month SMA'] = airline['Thousands of Passengers'].rolling(window=6).mean()
airline['12month SMA'] = airline['Thousands of Passengers'].rolling(window=12).mean()
```

```python
airline[['6month SMA','12month SMA']]
```

```python
airline.plot(figsize = (12,8))
```

```python
# 오차구하기
# 실제값 - 예측값

airline['Thousands of Passengers'] - airline['6month SMA']
```

```python
# 모델 평가하기

from sklearn.metrics import mean_squared_error

## 결측값제외하고 mean_squared_error(y값,예측y값)
RMSE = mean_squared_error(airline['Thousands of Passengers'].iloc[5:],airline['6month SMA'].dropna())
print(RMSE**0.5)
```

#### Weighted  Moving Average

* 가중이동평균
* window size가 커질수록 트렌드를 제대로 반영하지못하고, 노이즈,극단값의 영향을 크게 받는다 
* 최근의 값일수록 가중치를 크게주는 가중이동평균

```python
# ewm()이용 가중이동평균 구하기

airline['WMA12'] = airline['Thousands of Passengers'].ewm(span=12).mean() # window=12 가중이동평균
airline['WMA6'] = airline['Thousands of Passengers'].ewm(span=6).mean() # window=6 가중이동평균
```

```python
airline[['Thousands of Passengers','WMA6','WMA12']]
```

```python
airline[['Thousands of Passengers','WMA6','WMA12']].plot(figsize=(12,8))
```

> SMA의 경우 window가 길수록 트렌트만 따라가는(계절성등을 잘 반영하지못하는) 경향이 강했으나, WMA이는 이에반해 조금 더 트렌드와 특성을 반영하는 부분

```python
# 모델 평가하기

## 결측값제외하고 mean_squared_error(y값,예측y값)
RMSE = mean_squared_error(airline['Thousands of Passengers'],airline['WMA12'])
print(RMSE**0.5)
```

#### Simple Exponential Smoothing

* 단순지수평활법
* t+1시점의 예측값 = t시점의 예측값 + a(t시점의 실제예측값 - t시점의 예측값)     ( 0 <= a <= 1)

> 1. Simple Exponential Smoothing => 트렌드와 계절성을 반영 못함
> 2. Holt's => 트렌드만 반영
> 3. Holt-Winter's => 트렌드와 계절성까지 반영해줌

#### SES 1. Simple Exponential Smoothing

```python
from statsmodels.tsa.api import SimpleExpSmoothing
```

```python
# 모델에 fitting하기위한 작업 필요

train = airline[:'1995']
test = airline['1960':]
```

```python
train['Thousands of Passengers'].plot(figsize=(15,6))
test['Thousands of Passengers'].plot()
```

```python
ses_model = SimpleExpSmoothing(np.asarray(train['Thousands of Passengers'])) # numpy 배열로 변환하여
```

```python
ses_result = ses_model.fit()
```

```python
y_hat = test.copy() # 원본 test데이터 오염방지위해)
```

```python
y_hat['SES'] = ses_result.forecast(len(test)) #인자로 몇개를 예측할것인지
```

```python
plt.figure(figsize=(15,8))
plt.plot(train['Thousands of Passengers'],label='Train')
plt.plot(test['Thousands of Passengers'],label='Test')
plt.plot(y_hat['SES'],label='Simple Exp Smoothing')
plt.legend()
```

```python
# 모델 평가하기

RMSE = mean_squared_error(test['Thousands of Passengers'],y_hat['SES'])
print(RMSE**0.5)
```

#### SES 2.  Holt's model

```python
from statsmodels.tsa.api import Holt
```

```python
holt_model = Holt(np.asarray(train['Thousands of Passengers']))
```

```python
holt_result = holt_model.fit()
```

```python
y_hat['HOLT'] = holt_result.forecast(len(test))
```

```python
plt.figure(figsize=(15,5))
plt.plot(train['Thousands of Passengers'], label='Train')
plt.plot(train['Thousands of Passengers'], label='Test')
plt.plot(y_hat['HOLT'], label='Holts Exp Smoothing')
plt.legend()
```

```python
# 모델 평가하기

RMSE = mean_squared_error(test['Thousands of Passengers'],y_hat['HOLT'])
print(RMSE**0.5)
```

#### SES 3. Holt-Winter's 

```python
from statsmodels.tsa.api import ExponentialSmoothing
```

```python
winter_model = ExponentialSmoothing(np.asarray(train['Thousands of Passengers']),
                                   seasonal_periods=12, # 인자로 계절성파악위해 주기 주어야함(12개월 주기)
                                    trend='add', # addtive
                                    seasonal='add' # addtive
                                   ) 
```

```python
winter_result = winter_model.fit()
```

```python
y_hat['WINTER'] = winter_result.forecast(len(test))
```

```python
plt.figure(figsize=(15,5))
plt.plot(train['Thousands of Passengers'], label='Train')
plt.plot(train['Thousands of Passengers'], label='Test')
plt.plot(y_hat['WINTER'], label='HOLT-WINTER')
plt.legend()
```

```python
# 모델 평가하기

RMSE = mean_squared_error(test['Thousands of Passengers'],y_hat['WINTER'])
print(RMSE**0.5)
```
