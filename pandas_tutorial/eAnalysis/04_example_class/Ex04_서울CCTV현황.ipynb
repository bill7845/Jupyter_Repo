## 1. 서울시 자치구 연도별 CCTV 현황

```python
# 1. 데이타 살펴보기

import pandas as pd
import numpy as np

""" 1. 서울시 자치구 연도별 CCTV 현황 """
cctvSeoul = pd.read_csv('./data/cctv_in_seoul.csv',encoding='utf-8')
cctvSeoul.head()
# (1) CCTV가 많은 순으로 정렬하여 상위 5위 



# (2) 2014~2016년 3년간 CCTV 증가율을 추가
#     3년동안의 증가율 = ( 2014년 + 2015년 + 2016년 ) / 2013년도 이전 * 100
#     3년간 증가율 높은 순으로 출력
```

```python
cctvSeoul.sort_values(ascending=False,by='소계').iloc[:5]
```

```python
def func_1(x):
    result = (x[1]+x[2]+x[3]) / x[0] * 100
    return result

cctvSeoul['3년간 증가율'] = cctvSeoul[["2013년도 이전","2014년","2015년","2016년"]].apply(func_1,axis=1)
cctvSeoul.rename(columns={cctvSeoul.columns[0] :"구별"}, inplace = True)
cctvSeoul.sort_values(by='3년간 증가율',ascending=False)
```
